```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def aaa():
  f = lambda x: 2**(x/100)/1000

  sleeps = np.vectorize(f)(np.arange(1, 1001))
  print("  1", sleeps[0])
  print(" 10", sleeps[9])
  print("100", sleeps[99])
  print("500", sleeps[499])

  cum_sleeps = np.cumsum(sleeps)
  print("  1", cum_sleeps[0])
  print(" 10", cum_sleeps[9])
  print(" 20", cum_sleeps[19])
  print(" 50", cum_sleeps[49])
  print("100", cum_sleeps[99])
  print("150", cum_sleeps[149])
  print("200", cum_sleeps[199])
  print("500", cum_sleeps[499])
  print("1000", cum_sleeps[999])

  ratio = sleeps * np.vectorize(lambda x: 1/x)(cum_sleeps)
  print("  1", f"{ratio[0]*100:.1f}%")
  print(" 10", f"{ratio[9]*100:.1f}%")
  print(" 20", f"{ratio[19]*100:.1f}%")
  print(" 50", f"{ratio[49]*100:.1f}%")
  print("100", f"{ratio[99]*100:.1f}%")
  print("150", f"{ratio[149]*100:.1f}%")
  print("200", f"{ratio[199]*100:.1f}%")
  print("500", f"{ratio[199]*100:.1f}%")

  plt.plot(sleeps)
  plt.show()

  plt.plot(cum_sleeps)
  plt.show()

  plt.plot(ratio[5:])
  plt.show()

```

```python
aaa()
```
